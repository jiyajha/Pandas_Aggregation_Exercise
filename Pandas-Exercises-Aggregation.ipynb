# Aggregations

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset occupation.csv from the folder or [here](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
path = "Exercise Data Files/Pandas Exercise Aggregation files/"
```

```python
users = pd.read_csv(path + "occupation.csv", sep = "|")
```

### Step 3. Assign it to a variable called users.

```python
users.head(2)
```

### Step 4. Discover what is the mean age per occupation

```python
users.pivot_table(index = "occupation", values = "age", aggfunc = "mean")
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
result = users.pivot_table(index = "occupation", columns = "gender", aggfunc = "size", fill_value = 0)
```

```python
sums = result[["M", "F"]].sum(axis = 1)
```

```python
result["Male Ratio"] = result["M"]/sums
```

```python
result["Male Ratio"].sort_values(ascending = False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby("occupation")["age"].agg(["min", "max"])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(["occupation", "gender"])["age"].mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
result["Women Percentage"] = result["F"]/sums * 100
result["Men Percentage"] = result["M"]/sums * 100
```
